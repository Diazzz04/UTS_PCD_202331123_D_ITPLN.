```python
from PIL import Image, ImageEnhance, ImageOps
import matplotlib.pyplot as plt
```

# Membaca Gambar

```python
image_path = 'BL_UTS.jpg'
original_image = Image.open(image_path)
```

# Konversi ke Grayscale

```python
grayscale_image = ImageOps.grayscale(original_image)
```

# Grayscale yang di percerah

```python
only_bright = ImageEnhance.Brightness(grayscale_image).enhance(1.5)
```

# Grayscale yang di percontrass

```python
only_contrast = ImageEnhance.Contrast(grayscale_image).enhance(1.5)
```

# Grayscale yang di percerah dan contras

```python
bright_then_contrast = ImageEnhance.Contrast(only_bright).enhance(1.5)
```

# Tampilkan Gambar

```python

fig, axs = plt.subplots(2, 3, figsize=(24, 12))  # Ukuran diperbesar

# Gambar asli
axs[0, 0].imshow(original_image)
axs[0, 0].set_title("Gambar Asli" , fontsize=30 )
axs[0, 0].axis("off")

# Grayscale
axs[0, 1].imshow(grayscale_image, cmap='gray')
axs[0, 1].set_title("Grayscale",fontsize=30 )
axs[0, 1].axis("off")

# Grayscale + Brightness
axs[0, 2].imshow(only_bright, cmap='gray')
axs[0, 2].set_title("Grayscale + Brightness", fontsize=30)
axs[0, 2].axis("off")

# Grayscale + Contrast
axs[1, 0].imshow(only_contrast, cmap='gray')
axs[1, 0].set_title("Grayscale + Contrast", fontsize=30)
axs[1, 0].axis("off")

# Grayscale + Brightness + Contrast
axs[1, 1].imshow(bright_then_contrast, cmap='gray')
axs[1, 1].set_title("Grayscale + Brightness + Contrast",fontsize=30)
axs[1, 1].axis("off")

# Kosongkan slot terakhir
axs[1, 2].axis("off")

plt.tight_layout()
plt.show()

```
